```python
from langgraph.graph import StateGraph ,  START , END
from langchain_google_genai import ChatGoogleGenerativeAI
from typing import TypedDict
from dotenv import load_dotenv
```

```python
load_dotenv()
```

```python
model = ChatGoogleGenerativeAI(model="models/gemini-1.0-pro")
```

```python
#Create a State
class LLMStata(TypedDict):
    question: str
    answer: str
```

```python
def llm_qu(state: LLMStata):
    #extract the question from state
    question=state['question']

    #form a prompt
    prompt=f"Answer the following Questions {question}"

    #ask that question to the llm
    answer=model.invoke(prompt).content

    #update the answer in the state
    state["answer"]=answer

    
    return state
```

```python
#Create A graph
graph = StateGraph(LLMStata)
 
#add nodes
graph.add_node("llm_qu",llm_qu)

#Add edges
graph.add_edge(START,"llm_qu")

graph.add_edge("llm_qu",END)

workflow=graph.compile()

```

```python
initial_state={"question":"how far is the moon from the earth"}
final_state = workflow.invoke(initial_state)
print(final_state)
```
